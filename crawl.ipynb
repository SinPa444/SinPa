```python
import bs4 as bs
import requests
import pandas as pd
import csv
```

```python
url = "https://www.varzesh3.com/news/"
```

```python
title1_list = []
title2_list = []
views_list = []
tags_list = []
```

```python
for q in range(2086100,2087100):
    try:
        print(q)
        page = url + str(q) + "/"
        req = requests.get(page)
        html = bs.BeautifulSoup(req.content)
        title1 = html.find("h1", class_="headline").get_text().replace("\u200c"," ")
        title1_list.append(title1)
        title2 = html.find("p", class_="lead").text.replace("\n","").strip().replace("\u200c"," ").replace("\u200f"," ")
        title2_list.append(title2)
        tags = html.find("div", class_="tags tags-news").text.replace("\n","").strip().replace("\u200c"," ").replace("\u200f"," ")
        tags_list.append(tags)
        views = html.find("div", class_="news-info")
        view = int(float(views.find_all("span")[2].get_text().split(" ")[0][:-1]) * 1000)
        views_list.append(view)
        print("first_title:")
        print(title1)
        print("secend_title:")
        print(title2)
        print("views:")
        print(view)
        print("tags:")
        print(tags)
    except AttributeError:
        print("not a news")
```

```python
file_name = "crawl_info.csv"
```

```python
with open(file_name, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["title1_list", "title2_list", "views_list", "tags_list"])
    
    for t1, t2, v, t in zip(title1_list, title2_list, views_list, tags_list):
        writer.writerow([t1, t2, v, t])
```
